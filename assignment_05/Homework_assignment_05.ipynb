## 8. writing information about the column data

#### 1. UNITID (Unique identification number)
     - it is the unique identification number assigned to postsecondary institutions as surveyed through IPEDS.


#### 2. INSTNM (Institution name)
     - it is the name of the Institution as mentioned in the IPEDS


#### 3. CITY (place where the college is located at)
    - The Institutes location as reported in the IPEDS, this can also be reported using longitude (float value) or latitude 


#### 4. STABBR (location of institution)
     - This is also used to represesnt the institutes location 


#### 5. ZIP (unique code to identify location)
     - a code given to identify a particular location which is unique to every place


#### 6. ADM_RATE (Admission rate) 
     - Colleges report to IPEDS their Fall admissions rate defined as the number of admitted undergraduates divided by the number of undergraduates who applied. 
     -  For institutions with multiple branches, ADM_RATE includes the admissions rate at each campus, while
     - ADM_RATE_ALL represents the admissions rate across all campuses, defined as the total number of admitted undergraduates across all branches divided by the total number of undergraduates who applied across all branches. 

#### 7. UGDS (Undergraduate Student Body by Race and Gender)
     - These data are reported by institutions to IPEDS in the Fall Enrollment component; and rely on students’ self-reported race and gender data, as collected by the institution.
     - These include men (UGDS_MEN), women (UGDS_WOMEN) and so on

#### 8. TUITIONFEE_IN (In tuition fees)
    - the cost data includes the tuition and required fees, books and supplies, and living expenses which are estimated by the institution. 
    - Tuition and required fees are provided for in-state students.


#### 9. MN_EARN_WNE_P10 (Earnings)
     -  Mean and median earnings are for the institutional aggregate of all federally aided students who enroll in an institution each year and who are employed but not enrolled.
     - earnings are defined as the sum of wages and deferred compensation from all non-duplicate W-2 forms received for each individual
