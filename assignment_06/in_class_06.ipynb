```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('wdi_data.csv')
df.head()
```

```python
df = df[df['indicator']== 'SP.POP.TOTL']
```

```python
df
```

```python
df = df[df['Country Code']=='USA']
```

```python
df.head()
```

```python
fig = go.Figure()

trace_0 = go.Bar(
    x = df['Year'],
    y = df['value']
)

fig.add_trace(trace_0)
fig.show()
```

```python
fig.update_layout(
    title = 'USA Population growth',
    xaxis = {'title':'year'},
    yaxis = {'title':'population'}
)

fig.show()
```

```python
fig = go.Figure()

trace_0 = go.Scatter(
    x = df['Year'],
    y = df['value'],
    mode = 'markers'
    #mode = 'markers+lines'
)

fig.add_trace(trace_0)

fig.update_layout(
    title = 'USA Population growth',
    xaxis = {'title':'year'},
    yaxis = {'title':'population'}
)

fig.show()
```

```python
df_ind = pd.read_csv('wdi_data.csv')
```

```python
df_ind = df_ind[df_ind['indicator']== 'SP.POP.TOTL']
df_ind = df_ind[df_ind['Country Code']=='IND']
```

```python
fig = go.Figure()

trace_0 = go.Scatter(
    x = df_ind['Year'],
    y = df_ind['value'],
    #mode = 'markers'
    mode = 'markers+lines'
)

fig.add_trace(trace_0)

fig.update_layout(
    title = 'IND Population growth',
    xaxis = {'title':'year'},
    yaxis = {'title':'population'}
)

fig.show()
```
